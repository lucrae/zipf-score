```python
%matplotlib inline

import re
import math
from collections import Counter

import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
```

```python
from load import load_brown, load_random, load_bible

# load words
words = load_brown()
```

```python
# count and rank words
word_count = dict(Counter(words).most_common())

# print info on corpus
n_words = len(words)
n_unique_words = len(word_count)
word_table = [(i+1, word, f'{word_count[word]}') for i, word in enumerate(list(word_count)[:10])]
print(f'# Words: {n_words:,}')
print(f'# Unique words: {n_unique_words:,}')
print('')
print(tabulate(word_table, headers=['Rank', 'Word', 'Freq.']))
```

```python
# define zipf-mandelbrot formula
def zipf(r, c=1, alpha=1, beta=0):
    return c / (r + beta)**alpha
```

### Zipf's Law

![Zipf's Law](images/formulae/zipf.png)

Zipf's Law states that if the words in a given corpus or large text are ranked by frequency, the frequency of each word can be approximated as inversely proportional to its rank.

### Zipf-Mandelbrot Law

![Zipf-Mandelbrot Law](images/formulae/zipf-mandelbrot.png)

The Zipf-Mandelbrot generalisation allows closer fits by "shifting" the rank by an amount $\beta$.

```python
# calculate example zipf curve
z_x = np.arange(1, 100)
z_y = zipf(z_x, c=1)

# set up figure
fig = plt.subplots(figsize=(8, 4))
colors = [(1-val, 0, val) for val in z_x/np.max(z_x)]

# plot zipf sample scatter plot in linear space
plt.subplot(121)
plt.scatter(z_x, z_y, color=colors)
plt.title('Linear Space')
plt.xlabel('x', fontsize=12)
plt.ylabel('zipf(x)', fontsize=12)

# plot zipf sample scatter plot in log space
plt.subplot(122)
plt.scatter(np.log(z_x), np.log(z_y), color=colors)
plt.title('Log Space')
plt.xlabel('$log_e$ x', fontsize=12)
plt.ylabel('$log_e$ zipf(x)', fontsize=12)

# show plot
plt.tight_layout()
plt.show()
```

```python
# produce rank and frequency as np arrays
rank = np.array(range(1, n_unique_words+1))
frequency = np.array(list(word_count.values()))

# calculate log values
rank_log = np.log(rank)
frequency_log = np.log(frequency)

# calculate zipf score
prop_zipf = lambda r: zipf(r, c=np.max(frequency))
zipf_log = np.log(prop_zipf(rank))
```

```python
# calculate mean absolute error (MAE)
error_log = frequency_log - zipf_log
mean_absolute_error = np.sum(np.abs(error_log)) / len(error_log)

# calculate coefficient of determination (R^2)
mean_frequency_log = np.mean(frequency_log)
total_sum_of_squares = np.sum((frequency_log - mean_frequency_log)**2)
residual_sum_of_squares = np.sum((frequency_log - zipf_log)**2)
coefficient_of_determination = 1 - (residual_sum_of_squares / total_sum_of_squares)
```

```python
def plot(rank, frequency,
         rank_log, frequency_log,
         zipf_log, error_log,
         mean_absolute_error, coefficient_of_determination):
    '''Plots the distribution, log-space distribution, and error for given calculations'''

    # set up figure
    fig, axes = plt.subplots(1, 3, figsize=(12, 4))

    # scatter plot
    ax = axes[0]
    ax.scatter(rank, frequency, s=16, marker='.', color='#00aaaa')
    ax.set_title(f'Word Frequency (n={n_words:,})')
    ax.set_xlabel('Frequency Rank')
    ax.set_ylabel('Frequency')
    if np.min(frequency) != 1:
        ax.set_ylim(bottom=0)

    # scatter plot log-space
    ax = axes[1]
    ax.scatter(rank_log, frequency_log, s=16, marker='.', color='#00aaaa')
    ax.plot(rank_log, zipf_log, color='r', alpha=0.3, linestyle='--', linewidth=2,
            label='Power-Law Curve')
    ax.set_title('Log-Space Distribution')
    ax.set_xlabel('$log_e$ Frequency Rank')
    ax.set_ylabel('$log_e$ Frequency')
    ax.legend()

    # error
    ax = axes[2]
    ax.scatter(rank_log, error_log, s=16, marker='.', color='#aa66aa')
    ax.plot([0, np.max(rank_log)], [0, 0], color='r', alpha=0.3, linestyle='--', linewidth=2,
            label='Power-Law Curve')
    ax.set_ylim((-16, 16))
    ax.set_title('Log-Space Error')
    ax.set_xlabel('$log_e$ Frequency Rank')
    ax.set_ylabel('Log-Space Error')
    text_str = f'MAE = {mean_absolute_error:.3f}\n$R^2$ = {coefficient_of_determination:.3f}'
    ax.text(0, -6, text_str, fontsize=10, color='#aa00aa')
    ax.legend()

    # show plot
    plt.tight_layout()
    plt.show()
```

```python
# plot the distribution, log-space distribution, and log-space error
plot(rank, frequency, rank_log, frequency_log,
     zipf_log, error_log, mean_absolute_error, coefficient_of_determination)
```
